Part 1)

```python
import numpy as np; from scipy import stats; import matplotlib.pyplot as plt
```

```python
n = 5; x = stats.norm(loc=0,scale=1).rvs(n); theta_0, tau = 4, 1/2; parameter_domain = np.linspace(-10,10,1001); 
plt.figure(figsize=(16,5))
posterior = stats.norm(loc=(tau*theta_0+x.sum())/(tau+n), scale=1/(tau+n)**0.5).pdf(parameter_domain)
for xi in x:
    likelihood = stats.norm.pdf((x[:,np.newaxis])*np.ones(parameter_domain.shape), loc=parameter_domain, scale=1).prod(axis=0)
    plt.plot(parameter_domain, likelihood, label="likelihood "+ str(xi) +"")

plt.plot(parameter_domain, posterior, label="posterior"); plt.legend(); plt.title("Likelihoods"); 
```

Part 2)

```python
likelihood = stats.norm.logpdf((x[:,np.newaxis])*np.ones(parameter_domain.shape), loc=parameter_domain, scale=1).prod(axis=0)
prior = stats.norm(loc=theta_0, scale=1/tau**0.5).logpdf(parameter_domain) 
joint = likelihood*prior
posterior = stats.norm(loc=(tau*theta_0+x.sum())/(tau+n), scale=1/(tau+n)**0.5).logpdf(parameter_domain)

plt.figure(figsize=(16,5)); plt.plot(parameter_domain, likelihood, label="loglikelihood"); plt.plot(parameter_domain, prior, label="prior"); plt.plot(parameter_domain, posterior, label="posterior"); plt.plot(parameter_domain, joint, linestyle="--", label="joint"); plt.legend(); plt.title("Log normal normal"); 
```

Part 3) Posterior distribution of a normal likelihood with ()1 looks like a normal normal model or a normal distribution.

Part 4) For a log likelihood we can see that the prior does not integrate to 1 and the shape of the distribution for the loglihood is different from that of the the posterior, unlike in part 1) where the distribution of the likelihood and the posterior are the same shape.This may be due to the constant that is applied under log.
