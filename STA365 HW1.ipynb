```python
import numpy as np; from scipy import stats; import matplotlib.pyplot as plt
stats.binom(n=1,p=0.5).rvs(size=1)
#stats.bernoulli(p=0.5).rvs(size=1)
```

```python
import random; random.seed(1234)
p_a, p_b, p_c = 0.3, 0.5, 0.7

#A
a_A, b_A, n, x = 1, 1, 1, 1

for i in range(10):
    if stats.binom(n = n, p = p_a).rvs(size=1) == 1:
        x += 1
    n += 1
a_A = x + a_A - 1; b_A = n - x + b_A - 1
print(a_A)
print(b_A)
```

```python
#B
a_B, b_B, n, x = 1, 1, 1, 1

for i in range(10):
    if stats.binom(n = n, p = p_b).rvs(size=1) == 1:
        x += 1
    n += 1
a_B = x + a_B - 1; b_B = n - x + b_B - 1

print(a_B)
print(b_B)
```

```python
#C
a_C, b_C, n, x = 1, 1, 1, 1

for i in range(10):
    if stats.binom(n = n, p = p_c).rvs(size=1) == 1:
        x += 1
    n += 1
a_C = x + a_C - 1; b_C = n - x + b_C - 1
print(a_C)
print(b_C)
```

```python
parameter_domain = np.linspace(0,1,1001)
posterior_A = stats.beta(a = a_A, b = b_A ).pdf(parameter_domain)
posterior_B = stats.beta(a = a_B, b = b_B ).pdf(parameter_domain)
posterior_C = stats.beta(a = a_C, b = b_C ).pdf(parameter_domain)

plt.figure(figsize=(12,4)); plt.plot(parameter_domain, posterior_A, label="posterior A, p = "+str(p_a)+""); plt.plot(parameter_domain, posterior_B, label="posterior B, p = "+str(p_b)+""); plt.plot(parameter_domain, posterior_C, label="posterior C, p = "+str(p_c)+""); plt.legend(); plt.title("Plots of Posteriors");  
```

We see that the shape of the curve for posterior C has the highest point but is centered around 0.1. While posterior A and B have flatter plots with their peaks further towards 1. I believe that Posterior C would be the ideal choice and we can see that as p increases, the peaks of the posterior's distributions increases with the peaks shifting towards 0.
